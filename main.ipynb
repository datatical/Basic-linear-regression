```python
import pandas as pd
import matplotlib.pyplot as plt

data_df = pd.read_csv('Sheet1.csv')

plt.scatter(data_df['X'], data_df['Y'])
plt.xlabel('Y')
plt.ylabel('X')
plt.title('Scatter plot of X vs Y')
plt.show()

X = data_df['Y']
y = data_df['X']

n = len(X)
x_mean = X.mean()
y_mean = y.mean()
numerator = 0
denominator = 0
for i in range(n):
    numerator += (X[i] - x_mean) * (y[i] - y_mean)
    denominator += (X[i] - x_mean) ** 2
slope = numerator / denominator
intercept = y_mean - slope * x_mean

reg_line = slope * X + intercept
plt.scatter(X, y)
plt.plot(X, reg_line, color='red')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Regression line of X vs Y')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Read the data
data = pd.read_csv("Salary_Data.csv")
X = data.iloc[:, :-1].values
y = data.iloc[:, 1].values

# Split the data into training and test sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Train the model
regressor = LinearRegression()
regressor.fit(X_train, y_train)

# Predict the salaries for the test set
y_pred = regressor.predict(X_test)

# Plot the actual salaries in the test set and the predicted salaries
plt.scatter(X_test, y_test, color = 'red')
plt.plot(X_test, y_pred, color = 'blue')
plt.title('Salary vs Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()

# Print the slope and intercept of the linear regression line
print('Slope of the regression line is: ', regressor.coef_[0])
print('Intercept of the regression line is: ', regressor.intercept_)

# Print the accuracy of the model
print('Accuracy of the model is: ', regressor.score(X_test, y_test))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import statistics as stats
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from scipy.stats import shapiro

# Read the data
data = pd.read_csv("Sheet1.csv")
X = data.iloc[:, 0].values.reshape(-1, 1)
y = data.iloc[:, 1].values.reshape(-1, 1)

# Split the data into training and test sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Train the model
regressor = LinearRegression()
regressor.fit(X_train, y_train)

# Predict the output variable for the test set
y_pred = regressor.predict(X_test)

# Plot the actual output variable in the test set and the predicted output variable
plt.scatter(X_test, y_test, color='red')
plt.plot(X_test, y_pred, color='blue')
plt.title('X vs Y')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()

# Print the slope and intercept of the linear regression line
print('Slope of the regression line is:', regressor.coef_[0][0])
print('Intercept of the regression line is:', regressor.intercept_[0])

# Print the accuracy of the model
print('Accuracy of the model is:', regressor.score(X_test, y_test))

# Print the mean squared error of the model
print('Mean Squared Error of the model is:', mean_squared_error(y_test, y_pred))

# Check residual normality using Shapiro-Wilk Test
residuals = y_test - y_pred
stat, p = shapiro(residuals.flatten())
print('Residual Normality Test Result (p-value) is:', p)

# Print the mean of Y
print('Mean of Y is:', stats.mean(y_test.flatten()))
```
