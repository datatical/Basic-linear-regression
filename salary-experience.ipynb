```python

import pandas as pd
import matplotlib.pyplot as plt

salary_df = pd.read_csv('Salary_Data.csv')

plt.scatter(salary_df['YearsExperience'], salary_df['Salary'])
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Scatter plot of Years of Experience vs Salary')
plt.show()


X = salary_df['YearsExperience']
y = salary_df['Salary']

# Calculating the slope and y-intercept of the regression line
n = len(X)
x_mean = X.mean()
y_mean = y.mean()
numerator = 0
denominator = 0
for i in range(n):
    numerator += (X[i] - x_mean) * (y[i] - y_mean)
    denominator += (X[i] - x_mean) ** 2
slope = numerator / denominator
intercept = y_mean - slope * x_mean

# Plotting the regression line
reg_line = slope * X + intercept
plt.scatter(X, y)
plt.plot(X, reg_line, color='red')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Regression line of Years of Experience vs Salary')
plt.show()
```
